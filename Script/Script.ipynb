```python
import tabula
import os
```

```python
# Get the list of all files and directories
inpath = r"..\Rawdata"
indir_list = os.listdir(inpath)
indir_list[0]
infile= os.path.join(inpath, indir_list[0])
infile[:-4]
```

```python
# Get the list of all files and directories
outpath = r"..\Output"
outfile= os.path.join(outpath, indir_list[0])
outfile[:-4]
```

```python
inpdf = infile[:-4]+".pdf"
inpdf
```

```python
outcsv = outfile[:-4]+".csv"
outcsv
```

```python
# Read pdf into list of DataFrame
dfs = tabula.read_pdf(infile, pages='all')
```

```python
dfs
```

```python
# convert PDF into CSV file
tabula.convert_into(inpdf, outcsv, output_format="csv", pages='all')
```
